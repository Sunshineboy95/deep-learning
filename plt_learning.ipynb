```python
import random

x =range(60)
y=[random.uniform(15,20) for i in x]


#创建画布
plt.figure(figsize=(12,8),dpi=80)
plt.plot(x,y)

#设置刻度
z = range(40)
x_label=['11:{}'.format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(z[::5])

#添加网格
plt.grid(True, linestyle='--',alpha=0.5)

#添加标题
plt.xlabel('time')
plt.ylabel('temperature')
plt.title('deep learning')


#保存图片
plt.savefig('./test2.png')
plt.show()
```

```python
import random

x =range(60)
y1=[random.uniform(15,20) for i in x]
y2=[random.uniform(5,10) for i in x]


#创建画布
plt.figure(figsize=(12,8),dpi=80)
plt.plot(x,y1,color='r',linestyle='--',label='shanghai')
plt.plot(x,y2,color='g',linestyle='-.',label='shanghai')

#设置刻度
z = range(40)
x_label=['11:{}'.format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(z[::5])

#添加网格
plt.grid(True, linestyle='--',alpha=0.5)

#添加标题
plt.xlabel('time')
plt.ylabel('temperature')
plt.title('deep learning')


#保存图片
plt.savefig('./test3.png')
plt.show()
```
